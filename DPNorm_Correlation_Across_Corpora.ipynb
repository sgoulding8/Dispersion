```python
import pandas as pd
import collections
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from scipy.stats import pearsonr
```

```python
# Read in the data from the Excel file
dispListLara = pd.read_excel('dispersionLara.xlsx')
dispListSod = pd.read_excel('dispersionSod.xlsx')
dispListSam = pd.read_excel('dispersionSam.xlsx')
```

```python
# Merge the dataframes dispListLara, dispListSod, and dispListSam based on the common column 'Words'. 
# This creates a dataframe with only the words all three corpora have in common
common_words = pd.merge(dispListLara, dispListSod, on='Words', how='inner')
common_words = pd.merge(common_words, dispListSam, on='Words', how='inner')

# Select the required columns
common_words = common_words[['Words', 'DPNorm_x', 'DPNorm_y', 'DpNorm']]
common_words.columns = ['Words', 'DPNorm_Lara', 'DPNorm_Sod', 'DPNorm_Sam']

# # Print the common_words dataframe
print(common_words)
```

```python
dpnorm_scores = common_words[['DPNorm_Lara', 'DPNorm_Sod', 'DPNorm_Sam']]

# Calculate the correlation matrix of DPNorm across all three corpora
correlation_matrix = dpnorm_scores.corr()

# Print the correlation matrix
print(correlation_matrix)
```
